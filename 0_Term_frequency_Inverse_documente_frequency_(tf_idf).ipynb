#02. Conversão do pacote de texto em vatores:

```python
#Importar a biblioteca do panda:
tfidf = TfidfVectorizer(stop_words='english')
```

Para facilidade dos processos, as palavras de ligação são inseridas numa variável declarada bow.

```python
#Normalizando os dados:
tfidf.fit(text_data)
```

"Fitting" é parametrizar os dados. Por meio do método - ação - fit, deu-se a variável que amazena o banco "text_data" como argumetno da parametrização a fim de normalizá-la.

```python
#Retorna o índice das palavras mais significativas:
tfidf.vocabulary_
```

Esse recuro será útil para o entendimento do programa mais para frente.

```python
#Primeira transformação:
tfidf_features = tfidf.transform(text_data)
tfidf_features
```

Visto nas saídas acima, elas são retornadas numa linguagem mais computacional, a fim de compreensão, pode-se transformar esse resultado em algo mais compreensível. Para começar, armazenou-se a primeira transformação numa variável.

```python
#Segunda transformação:
tfidf_features_array = tfidf_features.toarray()
tfidf_features_array
```

Agora o pacote de texto foi efetivamente convertido numa array com valores vetoriais corresponde a significâncias das palavras que as compõem.

Retorna três conjuntos de dados:
1. As frases originais com seus elementos completos;
2. Uma tupla com duas ordens de elementos: os elementos nos índices zero - primeiros elementos -, são a posição que a frase ocupa no array text_data; o segundo é o índice que foi gerado para cada palavra significativa;
3. O valor vetorial dessa palavras.

#03. Similaridade de cossenos - Term Frequency - Inverse Document Frequency (tf-idf):

```python
#Teste:
#cos_sin = cosine_similarity(tfidf_features_array)
#lista_cos_sin = []
#for j in range(1,len(tfidf_features_array)):
#  lista_cos_sin.append(cos_sin[0,j])

#lista_cos_sin
```

```python
#Geração e armazenamento dos valores de similiraridade:
cos_sin = cosine_similarity(tfidf_features_array, [tfidf_features_array[0]])
lista_cos_sin = cos_sin.ravel().tolist()
lista_cos_sin
```

Através do método "Cosine_similarity - Similaridade de cosseno", possibilitou-se calcular os valores das similaridades. Lembre-se que, na segunda etapa, os valores da conversão do pacote de texto em vetores, foram amazenados numa variável declarada bow_feature_array. Aqui ela é trazida como argumento para o método cosine_similarity, para os valores serem resgatas e usados como recursos para o cálculo. Essa operação está armazenada em cos_sin. Em seguida, cria-se um laço que, enquanto estiver dentro do tamanho do pacote dde texto, calcula e armazenada os valores numa lista através da variável cos_sin - que contém o cálculo -. Lembre-se do demonstrado na teoria que o cálculo de similaridade de cossenos baseia-se numa comparação com a query, que, portanto, é constante para todas os demais elementos do pacote.

```python
#Geração de tuplas com as sinopses dos filmes:
sinopses_zipadas = list(zip(text_data,lista_cos_sin))
sinopses_zipadas
```

Usa-se a função zip para empacotar os dados em tuplas, esse é um recurso que nos permitirá de forma mais simples conseguir ordenar os resultados obtidos. As tuplas são geradas no seguinte padrão (text_data2[i],lista_cos_sin[i]), ou seja, iterado.

```python
#Ordenação dos resultados com as sinopses dos filmes:
sinopses_ordenadas = sorted(sinopses_zipadas, key = itemgetter(1), reverse=True)
sinopses_ordenadas
```

Como dito acima, a ordenação justifica a geração de tuplas. Através do "itemgetter - retorna/busca item", podemos determinar qual índice de elementos queremos trabalhar. O método "Sorted" - (organizador), é quem propriamente gera a ordenação por meio do seguintes método: de onde será extraída as informações para ordenação; um critério de ordenação, nessa caso, usou-se o itemgetter para determinar que o segundo índice, que armazena os cossenos, serão o critério; e a ordem da ordenação - crescente ou decrescente -.

```python
sinopses_filmes=[sinopses_ordenadas[0] for sinopses_ordenadas in sinopses_ordenadas]
sinopses_filmes.pop(0)
sinopses_filmes
```

Usou-se a compeensão de lista para pegar apenas os valores correspondentes a sinopses dos filmes. Antes eles estavam numa array, organizados em tuplas cujas continham (sinopses dos filmes, valor da similaridade de cossenos). O fatiamento sinopses_ordenadas[0], está selecionando apenas a informação no primeiro índice e armazenando na variável sinopses_filmes. Isso está sendo feito por toda a lista.

```python
#Geração de tuplas como os nomes dos filmes:
nomes_zipados = list(zip(names_data,lista_cos_sin[1:]))
nomes_zipados
```

```python
#Ordenação dos resultados com as sinopses dos filmes:
nomes_ordenados = sorted(nomes_zipados, key = itemgetter(1), reverse = True)
nomes_ordenados
```

```python
#Fatiando os nomes dos filmes:
nomes_filmes=[nomes_ordenados[0] for nomes_ordenados in nomes_ordenados]
nomes_filmes
```

```python
#Fatiamento das similaridades de cossenos:
sim_ordenadas = sorted(lista_cos_sin, reverse=True)
sim_ordenadas.pop(0)
sim_ordenadas
```
