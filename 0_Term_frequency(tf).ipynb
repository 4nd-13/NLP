O que o programa faz? Ele vai comparar sempre o que está na primeira posição [0], com os demais índices, de tal forma que, faz-se necessário, para uma semelhança de cossenos que, a query seja o primeiro índice do array, enquanto que, os documents, os demais.

```python
#Inserção dos dados da query e dos documentos num único "pacote" de texto:
text_data2 = []
text_data2.extend(query)

if text_data2[0] == None:
  print('Operação inválida: nenhum recurso de busca foi inserido')

elif text_data2[0] != None:
  text_data2.extend(dataset_data)
```

#02. Conversão do pacote de texto em vetores:

```python
#Seleção de stop word:
bow = CountVectorizer(stop_words='english')
```

Para facilidade dos processos, as palavras de ligação são inseridas numa variável declarada bow.

```python
#Normalizando os dados:
bow.fit(text_data2)
```

"Fitting" é parametrizar os dados. Através do método - ação - fit, deu-se a variável que armazena o banco "text_data2" como argumento da parametrização a fim de normalizá-los.

```python
#Retornando o resultado do filtro:
bow.get_feature_names_out()
```

A partir do método "Get feature names out - Retorne o filtro dos nomes", retona-se o resultado filtro ao qual o pacote foi submetido para se extrair as palavras menos significativas dele - palavras de ligação -.

```python
#Primeira transformação:
bow_features = bow.transform(text_data2)
bow_features
```

Visto nas saídas acima, elas são retornadas numa linguagem mais computacional, a fim de compreensão, pode-se transformar esse resultado em algo mais compreensível. Para começar, armazenou-se a primeira transformação numa variável.

```python
#Segunda transformação:
bow_features_array = bow_features.toarray()
bow_features_array
```

Agora o pacote de texto foi efetivamente convertido numa array com valores vetoriais corresponde a significâncias das palavras que as compõem.

```python
#Geração e armazenamento dos valores de similaridade:
cos_sin = cosine_similarity(bow_features_array, [bow_features_array[0]])
lista_cos_sin = cos_sin.ravel().tolist()
lista_cos_sin
```

Através do método "Cosine_similarity - Similaridade de cosseno", possibilitou-se calcular os valores das similaridades. Lembre-se que, na segunda etapa, os valores da conversão do pacote de texto em vetores, foram amazenados numa variável declarada bow_feature_array. Aqui ela é trazida como argumento para o método cosine_similarity, para os valores serem resgatas e usados como recursos para o cálculo. O método cosine_similarity já permite fazer o cálculo comparando dois valores, para isso deu-se dois argumento de bow_features_array - onde estão armazenados -, por meio do fatiamente [1:] - a partir de segundo elemento até o último - e [0] apenas o primeiro. Tudo já está sendo armazenado na lista cos_sin. Naturalmente o cos_sin se torna uma lista que contém lista, para corrigi-lo, usá-se o ravel, que pega apenas os valores dentro de listas, e tolist, que está alocando dentro de uma lista.

Usa-se a função zip para empacotar os dados em tuplas, esse é um recurso que nos permitirá de forma mais simples conseguir ordenar os resultados obtidos. As tuplas são geradas no seguinte padrão (text_data2[i],lista_cos_sin[i]), ou seja, iterado.

```python
#Geração de tuplas com as sinopses dos filmes:
sinopses_zipadas = list(zip(text_data2,lista_cos_sin))
sinopses_zipadas
```

Como dito acima, a ordenação justifica a geração de tuplas. Através do "itemgetter - retorna/busca item", podemos determinar qual índice de elementos queremos trabalhar. O método "Sorted" - (organizador), é quem propriamente gera a ordenação por meio do seguintes método: de onde será extraída as informações para ordenação; um critério de ordenação, nessa caso, usou-se o itemgetter para determinar que o segundo índice, que armazena os cossenos, serão o critério; e a ordem da ordenação - crescente ou decrescente -.  

```python
#Ordenação dos resultados com as sinopses dos filmes:
sinopses_ordenadas = sorted(sinopses_zipadas, key = itemgetter(1), reverse=True)
sinopses_ordenadas
```

Usou-se a compeensão de lista para pegar apenas os valores correspondentes a sinopses dos filmes. Antes eles estavam numa array, organizados em tuplas cujas continham (sinopses dos filmes, valor da similaridade de cossenos). O fatiamento sinopses_ordenadas[0], está selecionando apenas a informação no primeiro índice e armazenando na variável sinopses_filmes. Isso está sendo feito por toda a lista.

```python
#Fatiando as sinopses dos filmes:
sinopses_filmes=[sinopses_ordenadas[0] for sinopses_ordenadas in sinopses_ordenadas]
sinopses_filmes.pop(0)
sinopses_filmes
```

Os mesmos processos anteriores são aplicados para os nomes dos filmes.

```python
#Geração de tuplas com os nomes do filmes:
nomes_zipados = list(zip(names_data,lista_cos_sin[1:]))
nomes_zipados
```

```python
#Ordenação dos resultados com as sinopses dos filmes:
nomes_ordenados = sorted(nomes_zipados, key = itemgetter(1), reverse=True)
nomes_ordenados
```

```python
#Fatiando os nomes dos filmes:
nomes_filmes=[nomes_ordenados[0] for nomes_ordenados in nomes_ordenados]
nomes_filmes
```

```python
#Fatiamento das similaridades de cossenos:
sim_ordenadas = sorted(lista_cos_sin, reverse=True)
sim_ordenadas.pop(0)
sim_ordenadas
```

#04.Resultado:
